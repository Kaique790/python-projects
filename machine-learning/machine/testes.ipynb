```python
import pandas as pd
from sklearn import tree
```

```python
df_treino = pd.DataFrame({
    "Dor no peito": ["N", "S", "N", "S", "N", "S"],
    "Peso": [70, 70, 80, 80, 90, 90],
    "Ataque Cardíaco": ["N", "S", "N", "S", "S", "S"]
})

df_treino
```

```python
model = tree.DecisionTreeClassifier()
model
```

```python
X_treino = df_treino.iloc[:, :-1]
X_treino = X_treino.replace({"N": 0, "S": 1})

y_treino = df_treino.iloc[:, -1]

model.fit(X_treino, y_treino)
```

```python
model.feature_importances_
```

```python
tree.plot_tree(model)
```

```python
df_test = pd.DataFrame({
    "Dor no peito": ["N", "N", "S", "S"],
    "Peso": [65, 85, 75, 50],
    "Ataque Cardíaco": ["N", "S", "S", "S"]
})

x_test = df_test.iloc[:, :-1]
y_test = df_test.iloc[:, -1]

x_test = x_test.replace( { "S": 1, "N": 0 } )

y_pred = model.predict(x_test)
y_pred
```
